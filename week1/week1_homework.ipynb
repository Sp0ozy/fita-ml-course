## A. Problēmas apraksts

Gribam prognozēt, vai Titānika pasažieris izdzīvoja, balstoties uz viņa datiem. Tas palīdzētu saprast, kādi faktori ietekmēja izdzīvošanu. Noderētu kuģniecības kompāniju drošības dienestiem, lai uzlabotu evakuācijas procedūras.

## B. ML uzdevuma tips

Supervised learning, jo datos ir zināmā atbilde - kolonna survived.

Klasifikācija, jo prognozējam 0 vai 1.

## C. Target variable

```python
print(df['survived'].value_counts())
print(f'\nIzdzīvoja: {df["survived"].mean():.1%}')
```

Target mainīgais ir survived. 0 = neizdzīvoja, 1 = izdzīvoja. Aptuveni 38% pasažieru izdzīvoja.

## D. Features saraksts

```python
print('Skaitliskās features:')
print(df.select_dtypes(include='number').columns.tolist())

print('\nKategoriskās features:')
print(df.select_dtypes(include='category').columns.tolist())
```

Skaitliskās: age, fare, sibsp, parch

Kategoriskās: sex, pclass, embarked, class, alone

## E. Sākotnējie novērojumi

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(1)
print('Trūkstošās vērtības:')
print(pd.DataFrame({'Skaits': missing, '%': missing_pct})[missing > 0])
```

```python
print('Izdzīvošana pēc dzimuma:')
print(df.groupby('sex')['survived'].mean().apply(lambda x: f'{x:.1%}'))
```

Kolonnā age trūkst ~20% vērtību, deck kolonnā trūkst ~77% — to varētu izslēgt no modeļa.

Sievietes izdzīvoja daudz biežāk nekā vīrieši (~74% pret ~19%), tāpēc sex būs svarīga feature.
